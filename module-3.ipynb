1.	Masukkan iris.csv ke dataframe

```python
import pandas as pd
```

```python
df = pd.read_csv('..\..\..\..\..\iris.csv')
df
```

2.	Normalisasikan setiap attribut iris ke dataframe menggunakan normalisasi min max sehingga setiap nilai attribut memiliki range 0 – 1. Vbaru = (Vlama - Vmin) / (Vmax - Vmin). Vbaru = nilai data di suatu sel setelah normalisasi, Vlama = nilai data asli sebelum normalisasi, Vmax = nilai max dari kolom data tersebut, Vmin = nilai min dari kolom data tersebut       

```python
def normal(data):
    return (df[data] - df[data].min()) / (df[data].max() - df[data].min())

df["sepal.length"] = normal("sepal.length")
df["sepal.width"] = normal("sepal.width")
df["petal.length"] = normal("petal.length")
df["petal.width"] = normal("petal.width")
df
```

3.	Cari centroid dari data yang telah dinormalisasi. Centroid dihitung dari rata2 seluruh data suatu kolom

```python
cen_setosa = df.loc[df["variety"] == "Setosa"].mean()
cen_versicolor = df.loc[df["variety"] == "Versicolor"].mean()
cen_virginica = df.loc[df["variety"] == "Virginica"].mean()

print("=======Centroid Setosa========\n{} \n========Centroid versicolor========\n{} \n========Centroid virginica========\n{}".format(cen_setosa, cen_versicolor, cen_virginica))
```

4.	Masukkan iristesting.csv ke dataframe. Normalisasikan iristesting.csv menggunakan vmin dan vmax pada data 1

```python
dftest = pd.read_csv('..\..\..\..\..\iristesting.csv')
dftest

def normal(value):
    return (dftest[value] - dftest[value].min()) / (dftest[value].max() - dftest[value].min())

dftest['sepal.length'] = normal('sepal.length')
dftest['sepal.width'] = normal('sepal.width')
dftest['petal.length'] = normal('petal.length')
dftest['petal.width'] = normal('petal.width')
dftest
```

5.	Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv ke centroid masing2 class di iris. Gunakan euclidean distance 

```python
def euclid(indeks, value):
  dist = []
  for i in ['Setosa', 'Versicolor', 'Virginica']:
    cen_class = dftest[dftest['variety'] == i].mean()
    test = value.iloc[indeks, 0:4]
    a = ((test[0]-cen_class[0])**2 + (test[1]-cen_class[1])**2 + (test[2]-cen_class[2])**2 + (test[3]-cen_class[3])**2)**1/2
    dist.append(a)
  if (min(dist)== dist[0]):
    return 'Setosa'
  elif (min(dist)== dist[1]):
    return "Versicolor"
  elif (min(dist)== dist[2]):
    return "Virginica"
  else :
    return None

temp = []
for x in dftest.index:
  temp.append(euclid (x,dftest))
print(dftest['variety'] == temp)
print(dftest['variety'])

```

```python
print(f"Accuracy : {(dftest['variety'] == temp).sum()/len(dftest)*100}%")
```
